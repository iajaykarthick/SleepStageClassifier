```python
import os

import numpy as np
import matplotlib.pyplot as plt

from pyedflib import highlevel
from sleep_stage_classifier.config import DATA_DIR
```

```python
file_path = os.path.join(DATA_DIR, 'SC4001E0-PSG.edf')
hyp_file_path = os.path.join(DATA_DIR, 'SC4001EC-Hypnogram.edf')
```

```python
signals, signal_headers, header = highlevel.read_edf(file_path)
signal_labels = [sh['label'] for sh in signal_headers]
sampling_rates = [sh['sample_rate'] for sh in signal_headers]

fig, axs = plt.subplots(len(signals), 1, figsize=(12, 10), sharex=True)
for i, signal in enumerate(signals):
    times = np.arange(signal.size) / sampling_rates[i] 
    axs[i].plot(times, signal)
    axs[i].set_title(signal_labels[i])
    axs[i].title.set_size(8)

plt.xlabel('Time (s)')
plt.show()
```
