```python
import matplotlib.pyplot as plt
import numpy as np
import pyedflib
```

```python
def plot_signal(signal_data, sample_frequency, signal_label):
    import matplotlib.pyplot as plt
    import numpy as np

    # Create time vector
    time = np.arange(0, len(signal_data) / sample_frequency, 1 / sample_frequency)

    # Plot signal
    plt.figure()
    plt.plot(time, signal_data)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title(signal_label)
    plt.show()
```

```python
edf_file = pyedflib.EdfReader('../data/sleep-edf-database-expanded-1.0.0/sleep-cassette/SC4001E0-PSG.edf')

# Get the number of signals (channels) in the file
num_signals = edf_file.signals_in_file

# Get signal labels
signal_labels = edf_file.getSignalLabels()

for i in range(num_signals):
    # Get the sample frequency of the signal
    sample_frequency = edf_file.getSampleFrequency(i)

    # Get the signal data
    signal_data = edf_file.readSignal(i)

    # Plot the signal
    plot_signal(signal_data, sample_frequency, signal_labels[i])

# Close the EDF file
edf_file.close()
```
